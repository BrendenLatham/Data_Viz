```python
%%bigquery --project data-viz-338801
with w as (
  with q as (
    with t as (
      SELECT incident_group as Incident, hour_of_call as Hour
      FROM bigquery-public-data.london_fire_brigade.fire_brigade_service_calls
    )
    SELECT Incident, Hour
    FROM t
    GROUP BY Hour, Incident
    HAVING Hour BETWEEN 10 and 12 or Hour BETWEEN 1 and 4
  )
  SELECT *
  FROM q
  GROUP BY HOUR, Incident
  HAVING Incident = "False Alarm"
)
SELECT count(*) as FalseAlarmsFrom10to5
FROM w
```
