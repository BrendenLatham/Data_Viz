This assignment will be about encoding. First I will import my package and dataset

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

I will be looking at encoding the neighborhood_group and room_type columns

First I will see all the different values in those categories

```python
ng = pd.crosstab(index=df['neighbourhood_group'], columns='count') 
ng
```

```python
rt = pd.crosstab(index=df['room_type'], columns='count') 
rt
```

For the neighbourhood group I will make additional columns

```python
ng = df['neighbourhood_group']
ng = pd.get_dummies(ng.apply(pd.Series).stack()).groupby(level = 0).sum()
ng.head()
```

```python
df = pd.concat([df,ng],axis=1)
df.head()
```

Now for the room type I know it is more of a categorical type variable but for practice I am going to treat it as a ranking by which I would rather live in with 1 being the best and 3 the worst

```python
roomtype = {'Entire home/apt','Private room','Shared room'}
ranking = {1,2,3}
dictionary = dict({a:b for a,b in zip(roomtype,ranking)})
print(dictionary)
```

Now to add it to the main data table

```python
df.room_type = df.room_type.map(dictionary)
df.head()
```

There we have the data table in all it's encoded glory.
