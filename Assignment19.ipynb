For this assignment I will be working for NaNs

First I will import what I need

```python
import pandas as pd
import statistics as stat
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

Now I will be examing how many Null entries I have

```python
df.isnull().sum(axis = 0)
```

Well I can already tell you which columns I will be imputing

I want to see if the reason we have the Null reviews per month is becaus they don't have any reviews at all

```python
nulls = df[df['reviews_per_month'].isnull()]
nulls['number_of_reviews'].mean()
```

Because the mean number of reviews among all rows with Null reviews per month is 0, we can see that our hypothesis was correct

Now to change all the Null values to 0

```python
df['reviews_per_month'].fillna(value = 0)
```

```python
df.head()
```

Now for the last review column I will get the mean date

```python
dates = pd.DataFrame(df.last_review)
dates['last_review'] = pd.to_datetime(dates['last_review']).dt.date
dates['last_review'] = pd.to_datetime(dates.last_review, format = '%Y-%m-%d')
dates['last_review'].mean()
```

Now to replace the Nulls

```python
df['last_review'] = df['last_review'].fillna(value = '2018-10-04')
```
