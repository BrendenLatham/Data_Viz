Once again I will first import my packages and dataset

```python
import pandas as pd
import statistics as stat
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
```

```python
df.head()
```

I am going to be examining types. I will start with how the mean is effected when going from float to int

```python
print("as float: ",stat.mean(df.SepalLength))
print("as int: ", stat.mean(df.SepalLength.astype('int')))
```

It looks like it did change it by roughly .46

I want to see if the same happens when I group by Class

```python
df.groupby('Class').SepalLength.agg(['mean'])
```

```python
df2 = df
df2.SepalLength = df2.SepalLength.astype('int')
df2.groupby('Class').SepalLength.agg(['mean'])
```

It looks like the shift was in the same direction

Now I will gather the mean, median, count, and standard deviation of all columns when grouped by class

```python
df.groupby('Class').agg(['mean','median','count','std'])
```
