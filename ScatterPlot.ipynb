# Scatter Plots

In this section I will be looking at scatter plots with the workout dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

I will only be making one scatter plot, but I will include many variables

```python
ax = sns.scatterplot(data = df, x = 'calories', y = 'avg_heart_rate', hue = 'workout_type', style = 'workout_type', size = 'avg_cadence', zorder = 2)
ax.set(title = 'Calories VS Heart Rate With Approximate Average Activity Levels', xticks = [x for x in range(0,600,50)])
plt.legend(bbox_to_anchor=(1.25, 1), loc='upper left', borderaxespad=0)
sns.set(rc={'figure.figsize':(10,10)})

plt.axhspan(80, 100, facecolor= '0.8', alpha=0.3, zorder = 1)
plt.axhspan(100, 120, facecolor='0.6', alpha=0.3, zorder = 1)
plt.axhspan(120, 140, facecolor='0.4', alpha=0.3, zorder = 1)
plt.axhspan(140, 160, facecolor='0.2', alpha=0.3, zorder = 1)

plt.text(585, 90, "Moderate", horizontalalignment='left', size='medium', color='0.8', weight='semibold')
plt.text(585, 110, "Weight Control", horizontalalignment='left', size='medium', color='0.6', weight='semibold')
plt.text(585, 130, "Aerobic", horizontalalignment='left', size='medium', color='0.4', weight='semibold')
plt.text(585, 150, "Anaerobic", horizontalalignment='left', size='medium', color='0.2', weight='semibold')

plt.show()
```
