### Introduction

An ideal map of the kind I am exploring would have non-misleading data displayed in a manner that is easy to navigate by the common citizen. It would have clear borders seperating areas of interest, providing the ability to compare different location by how affected they are by the pandemic. In the next three sections I will act as a critic comparing maps by popular sources to the ideal.

### Fox

This map at first glance seems to be an example of effective data communication in a simple and concise way. Even having the option of splitting the view up between states and counties this map seems to do the job in the best strategy. However if one was to hover over the individual states, they would see a popup displaying the cumulatice case and death toles. If the data shown is then compared to either the 2021 or 2022 census, a different story starts to emerge. The color of the state in this map is dictated by nothing but the total cases reported for that state with no regard to population. This map quickly devolves to represent nothing more than the United States population distribution. When compared to a sensus it becomes plain to see that per capita Oklahoma is doing much worse that both Texas and California. This map is quick to make the areas with larger population look worse when it comes to fighting the spread of covid, however it tends to be the more medium sized populations that are having a worse time of it. This map is not used in an article, but rather pasted on it's own page on the Fox news website for everyone to see. It almost seems worse then that it is so misleading.
