# Visualization Conclusion

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/bank.csv')
df.head()
```

```python
df.balance = df.balance.astype('float')
def myfun(c):
  if c<0:
    c = c*-1
    c = math.log(c,10)
    c = c*-1
  elif c == 0:
    c = c
  else: 
    c = math.log(c,math.e)
  return c
df.balance = df.balance.apply(lambda x: myfun(x))
```

```python
ax = sns.scatterplot(data = df, x = 'age', y = 'balance', hue = 'job', style = 'education')
ax.set(title = 'Age VS Bank Balance Scaled By ln')
sns.set(rc={'figure.figsize':(10,10)})
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
```

```python
ax = df.balance.plot.hist(bins = 100)
ax.set(title = 'Bank Balance Scaled By ln')
```
