### Lesson 8

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

For this first webscraping section I will be getting a list of all Dallas Mavericks players.

```python
r = requests.get('https://en.wikipedia.org/wiki/Dallas_Mavericks')
html_contents = r.text
html_soup = BeautifulSoup(html_contents, 'lxml')
```

Now i'll examine the title of this page a bit.

```python
html_soup.title
```

Now for scraping the links from the team info table.

```python
links = html_soup.find("table",{"class":"toccolours"}).find_all('a', title=True)
MavsURL = []
for link in links:
  MavsURL.append('https://en.wikipedia.org/wiki/Dallas_Mavericks' + link['href'])
print(MavsURL)
```

### Lesson 9

Now I will process some data from https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films

```python
films = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
films_contents = films.text
FilmSoup = BeautifulSoup(films_contents, 'lxml')
film_tables = FilmSoup.find_all('table', class_= 'wikitable plainrowheaders')
film_tables[0].find_all('a')
FilmFrame = pd.read_html(str(film_tables))[0]
FilmFrame
```

```python
FilmFrame.columns = FilmFrame.columns.droplevel(-1)
```

```python
FilmFrame = FilmFrame.drop(columns='Unnamed: 5_level_0')
```

```python
FilmFrame = FilmFrame.drop([13])
FilmFrame
```
