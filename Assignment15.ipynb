# Cleaning Mountains

In this notebook I am going to be cleaning some data acraped from the web.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_highest_mountains_on_Earth')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
tables = html_soup.find_all('table',class_="wikitable")
df = pd.read_html(str(tables))[0]
df.columns = df.columns.droplevel(0).droplevel(0)
df.head()
```

First I will be cleaning up the column names

```python
df = df.rename(columns=lambda x: re.sub('[\(\[].*?[\)\]]', '', x.capitalize()))
df.head()
```

I am still not sure how I would go about cleaning the Country column but I figure it would be easier to put it into a string and work through it as a proof of concept.

```python
countries = re.sub('[\(\[].*?[\)\]]', '', str(df["Country "]))
countries = re.sub('[\n]', '', countries)
countries = re.sub('[\d-]','', countries)
countries = re.sub('Name: Country , Length: , dtype: object','', countries)
countries = re.sub('[^\w]+',',', countries)
countries = countries[1:]
countries = countries.replace(',', ' ') # I could totally do this in regex but I am sick of it
countries
```
