## Whiskers and Violins

For this assignment I will be making some box and whisker plots as well as some violin plots

```python
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid")
```

Now to import the data set

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

First I will create a boxplot for just calories

```python
ax = sns.boxplot(x=df['calories'])
```

Now I will examine the calories burned by each exercise

```python
ax = sns.boxplot(x=df['calories'],y=df['workout_type'])
```

It looks like all exercises burn a very similar number of calories

Now I will examine this data in the form of a violin plot

```python
ax = sns.violinplot(x=df['calories'])
```

As we can see, the violin plot is showing some negative values. We cannot burn negative calories in an exercise (unless the exercise is eating) so something weird is happening here.

The following is how i understand it, but might not be accurate. I would love some correction on this if it is wrong. A violin plot is used to describe estimated data density based on the data we have. We do have some data close to or equal to 0, so our violin plot by nature gives some slight probability that there will be outliers on that side. We know that isn't true.

Here is a more accurate violin plot going by the limits of our data

```python
ax = sns.violinplot(x=df['calories'], cut = 0)
```

```python
sns.violinplot(data = df, y = 'workout_type', x = 'calories', cut = 0)
```
